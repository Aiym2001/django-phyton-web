```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

Загрузка обучающей и тестовый выборок

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

Стандартизация выходных данных

```python
x_train = x_train / 255
x_test = x_test / 255
```

Преобразование выходныз значений в векторы по категориям

```python
y_train_cat = keras.utils.to_categorical(y_train, 10)
y_test_cat = keras.utils.to_categorical(y_test, 10)
```

отображение первых 25 изображений из обучающей выборки

```python
plt.figure(figsize=(10,5))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i], cmap=plt.cm.binary)

plt.show()
```

Формирование модели НС и вывод ее структуры в консоль

```python
model = keras.Sequential([
    Flatten(input_shape=(28, 28, 1)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

print(model.summary())      # вывод структуры НС в консоль
```

Компиляция НС с оптимизацией по ADAM и критерием - категориальная кросс-энтропия

```python
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

Запуск процесса обучения:80% - обучающая выборка, 20%-выборка валидации

```python
model.fit(x_train, y_train_cat, batch_size=32, epochs=5, validation_split=0.2)
```

```python
model.evaluate(x_test, y_test_cat)
```

Проверка распознавания цифр

```python
n = 1
x = np.expand_dims(x_test[n], axis=0)
res = model.predict(x)
print( res )
print( np.argmax(res) )

plt.imshow(x_test[n], cmap=plt.cm.binary)
plt.show()
```

Распознавание всей тестовой выборки

```python
pred = model.predict(x_test)
pred = np.argmax(pred, axis=1)

print(pred.shape)

print(pred[:20])
print(y_test[:20])
```
